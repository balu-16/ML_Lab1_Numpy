## 9) Time Series & Resampling (15m)

```python
tips_ts = tips.copy()
tips_ts['date'] = pd.to_datetime('2024-01-01') + pd.to_timedelta(np.arange(len(tips_ts)), unit='D')
tips_ts = tips_ts.set_index('date').sort_index()
weekly = tips_ts.resample('W')[['total_bill','tip']].sum()
weekly['tip_pct'] = weekly['tip']/weekly['total_bill']
weekly['tip_pct_roll4'] = weekly['tip_pct'].rolling(4, min_periods=1).mean()
weekly.head()
```

### ‚úçÔ∏è Analysis (Time Series & Resampling) ‚Äî 3‚Äì5 sentences
- What trend do you see weekly vs monthly?
  - Weekly total bill and tip amounts fluctuate quite a bit, while the rolling 4-week average tip percentage smooths out these variations, suggesting a more stable underlying tipping behavior over time.
- Interpret divergence between raw and rolling series.
  - The raw weekly series shows sharp peaks and valleys, while the rolling average is much smoother, indicating that the rolling average is capturing the general trend by reducing the impact of short-term fluctuations.
- Why pick `W` vs `MS`?
  - Using 'W' (weekly) resampling reveals more granular patterns and fluctuations compared to 'MS' (monthly start), which would further smooth out the data and potentially hide interesting weekly variations in tipping behavior.

## 10) Rolling & Window Ops (10m)

```python
daily = tips_ts.resample('D')[['total_bill']].sum()
daily['roll14_med'] = daily['total_bill'].rolling(14, min_periods=1).median()
daily[['total_bill','roll14_med']].head()
```

### ‚úçÔ∏è Analysis (Rolling & Window Ops) ‚Äî 3‚Äì5 sentences
- Why is median sometimes preferable to mean in rolling stats?
  - Median is preferable to mean in rolling statistics when the data contains outliers or is skewed. The median is less affected by extreme values than the mean, providing a more robust measure of the central tendency of the rolling window.
- What happens at the edges for rolling windows?
  - At the edges of the time series, the rolling window will have fewer data points than the specified window size. `min_periods` controls how many non-NaN values are required in the window to produce a non-NaN result. If the number of non-NaN values is less than `min_periods`, the result will be NaN.
- When is `expanding` more suitable than `rolling`?
  - `expanding` is more suitable than `rolling` when you want to calculate a statistic over all data points up to the current point in time. This is useful for cumulative calculations or when you want to see how a statistic changes as more data becomes available. `rolling` uses a fixed window size, which is more appropriate for analyzing local trends or smoothing short-term fluctuations.

### ‚úçÔ∏è Analysis (Dtypes & Memory) ‚Äî 3‚Äì5 sentences
- What memory savings did `category` or `string` yield?
  - `category` yielded significant memory savings for columns with a limited number of unique values (like 'sex', 'day', 'time'). `string` generally uses more memory than `object` for strings but provides better string-specific functionality.
- When would you avoid `category`?
  - Avoid `category` for columns with a very large number of unique values or when the order of categories is not meaningful, as it can increase memory usage and slow down operations in those cases.
- Any effect on joins/groupbys?

  - `cut` creates bins of equal width based on the value range of the data, while `qcut` creates bins with roughly equal numbers of observations based on quantiles. ## A) Data Cleaning & Feature Engineering

```python
df = titanic.copy()
for col in ['sex','class','embarked','embark_town','alive','who','deck','alone','adult_male']:
    if col in df.columns: df[col] = df[col].astype('category')
age_med = df.groupby(['sex','class'])['age'].transform('median')
df['age'] = df['age'].fillna(age_med)
if df['embark_town'].isna().any():
    mode_town = df['embark_town'].mode(dropna=True)
    if not mode_town.empty: df['embark_town'] = df['embark_town'].fillna(mode_town.iloc[0])
df['family_size'] = df['sibsp'].fillna(0) + df['parch'].fillna(0) + 1
df['fare_pp'] = df['fare'] / df['family_size']
df['deck'] = df['deck'].cat.add_categories(['Unknown']).fillna('Unknown')
bins = [0,12,18,35,50,80]; labels = ['Child','Teen','YoungAdult','MidAge','Senior']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, include_lowest=True)
df.head()
```
